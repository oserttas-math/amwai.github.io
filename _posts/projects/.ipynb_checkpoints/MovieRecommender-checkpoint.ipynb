## Recommender Systems

Recommender systems, also known as recommendation engines, provide for a more personalized web experience, be it on social media, shopping websites, or content providers such as Pandora and Netflix. There are two main types of recommender systems. There are those that employ content-based filtering and those that involve collaborative filtering. Content-based filters provide recommendations based on item characteristics. An example is where a recommendation for an item is made based on similar items that a user has utilized before. On the other hand, Collaborative Filtering (CF) involves providing recommendations based on how users have responded to an item.
Collabrative filtering can be further broken down to memory-based collaborative filtering and model-based collaborative filtering. As the name suggests, memory-based collaborative filtering involves building a model from the full dataset. The method algorithms utilize similarity metrics such as Euclidean distance, Pearson correlation, and cosine similarity to calculate user-item matches. In contrast, model-based collaborative filtering employs algorithms that extract information from the dataset that is then used to build a model to predict a user's rating for an item. In general, model-based CF tends to be less resource intense and better performing than memory-based CF algorithms, albeit harder to implement.
In this analysis, I will be exploring the use of model-based CF to build a movie-recommendation engine based on the 100k movie ratings MovieLens dataset.

## Model-based collaborative filtering

The model-based CF algorithm used in this analysis involves matrix factorization of the user-item matrix. The idea is to characterize the original matrix as the product of two low-rank matrices, and use these to discover latent features that influence the user-item interaction. When trying to find latent features, an assumption is made that the number of latent features is smaller than the number of users, as well as the number of items. This makes intituitive sense since it is the commonality of features between users that allows us to be able to recommend items to one user based on the ratings of another.

```python
#Get the total number of users as well as number of movies
num_users = df.user_id.max()
num_movies = df.movie_id.max()
num_users, num_movies
```

```python
#Obtain the training and test datasets using cross-validation
from sklearn import cross_validation as cv
train_data, test_data = cv.train_test_split(df, test_size=0.25)
```

```python
#Create a training and test matrix
#train_matrix = sparse.dok_matrix((num_users, num_movies), dtype=float)
train_matrix = np.zeros((num_users, num_movies))
for line in train_data.itertuples():
    user = line[1]
    movie = line[2]
    rating = line[3]
    if user > num_users:
        continue
    if movie > num_movies:
        continue
    if rating != 0:
        train_matrix[user-1, movie-1] = rating

#test_matrix = sparse.dok_matrix((num_users, num_movies), dtype=float)
test_matrix = np.zeros((num_users, num_movies))
for line in test_data.itertuples():
    user = line[1]
    movie = line[2]
    rating = line[3]
    if user > num_users:
        continue
    if movie > num_movies:
        continue
    if rating != 0:
        test_matrix[user-1, movie-1] = rating
```

```python
#Initialize the various variables
#K are the latent features, lmbda is the regularlization
#parameter lambda and "iterations" denotes the number of iterations
#to perform to convergence
R = train_matrix
T = test_matrix

M, N = R.shape
K = 3
lmbda = 0.1
iterations = 50
P = 5*np.random.rand(M, K)
Q = 5*np.random.rand(N, K)
```

```python
#This function calculates the RMSE. A penalty is imposed where we have an estimated rating
#but no actual rating in R. The penalty is imposed by multiplying by W
def calcRMSE(R, P, Q, W):
    rmse = np.sqrt(np.sum(pow((W*(R - np.dot(P, Q.T))), 2))/len(R[R>0]))
    return rmse    
```

To obtain the update rules, we take the derivatives of P and Q, and alternately update each factor as we fix the other. This is known as **Alternate Least Squares**. To avoid overfitting, a regularization term is introduced to the square error where lambda is the regularization parameter.

```python
#Establish update rules
def updateRules(R, T, P, Q, K):
    
    #Get the training weight matrix which will contain a 1 if a movie was rated and 0 if not
    W = R.copy()
    W[R > 0] = 1
    W[R == 0] = 0
    
    #Get the test weight matrix
    Z = T.copy()
    Z[T > 0] = 1
    Z[T == 0] = 0
    
    errors_T = [] #To store training RMSE
    errors_TT = [] #To store test RMSE
    E = np.eye(K) # This is a K x K identity matrix
    
    for i in xrange(iterations):
        
        #Update for P with fixed Q
        for u, Wu in enumerate(W): #Wu corresponds to the ratings by each user
            num_mov = np.count_nonzero(Wu) #Count of how many movies each user rated
            if num_mov == 0:
                num_mov = 1 # This is to guard against zero counts which would make the matrices to be singular
            #The least squares solution
            #Note that np.diag(Wu) takes the binary representation of each user (1, 0 if rated or not rated)
            #and converts into a diagonal array
            matrix = np.dot(Q.T, np.dot(np.diag(Wu), Q)) + lmbda*num_mov*E
            vector = np.dot(Q.T, np.dot(np.diag(Wu), R[u].T))
            #Factorize the matrix using LU decomposition, and then solve for P using
            #forward and backward substitution
            P[u,:] = np.linalg.solve(matrix, vector)
        
        #Update for Q with fixed P. Steps are duplicated
        for v, Wv in enumerate(W.T):
            num_users = np.count_nonzero(Wv) #Count of number of users that rated movie v
            if num_users == 0:
                num_users = 1
            matrix = np.dot(P.T, (np.dot(np.diag(Wv), P))) + lmbda*num_users*E
            vector = np.dot(P.T, np.dot(np.diag(Wv), R[:,v]))
            Q[v,:] = np.linalg.solve(matrix, vector)
    
        #Calculate RMSE for both the training and the test data
        #note that we use the P and Q trained on the training data
        #to test for generalization using the test data
        errors_T.append(calcRMSE(R, P, Q, W))
        errors_TT.append(calcRMSE(T, P, Q, Z))
        
    return P, Q, errors_T, errors_TT
```

```python
#Obtain the training and test errors
P_pr, Q_pr, training_err, test_err = updateRules(R, T, P, Q, K)
```

```python
#Plot the RMSE for both the train and test data as a function of iterations
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(range(iterations), training_err, marker='o', color="blue", label='Training Data');
plt.plot(range(iterations), test_err, marker='s', color="red", label='Test Data');
plt.legend(loc='upper right', frameon=False)
plt.xlabel('Iterations')
plt.ylabel('RMSE')
plt.show()
```

```python
#Obtain the prediction matrix
R_pred = np.dot(P_pr, Q_pr.T)
R_pred.shape
```

```python
#Re-scale the prediction matrix to the range 0-5
oldrange = (np.max(R_pred) - np.min(R_pred))
newrange = 5
R_pred = (((R_pred - np.min(R_pred)) * newrange) / oldrange) + 0
```
